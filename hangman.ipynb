```python
import random
#from hangman_words import word_list
#from hangman_art import stages

#create starting point with 6 lives
lives = 6

#from hangman_art import logo

#have a word randomly selected from the python hangman word list
chosen_word = random.choice(word_list)
print(chosen_word)

placeholder = ""
word_length = len(chosen_word)
print(logo)
for position in range(word_length):
    placeholder += "_"
print("Word to guess: " + placeholder)

display = placeholder
guessed_letters = []


#Create a While Loop to have the user guess a letter. Create if, elif, else statements for user's guesses to be confirmed
# or rejected. If rejected, user should lose a life and the correct stage of hangman should be displayed. If confirmed,
#letter should be added to the "display". User is then to be asked for the next letter until lives are spent or word is guessed.
while display != chosen_word:
    print(lives, "LIVES LEFT")
    guess = input("Guess a letter: ").lower()

    if guess in guessed_letters:
        print("you have guessed", guess,"already")
    else:
        if guess in chosen_word:
            new_display = ""
            for i in range(len(chosen_word)):
                if chosen_word[i] == guess:
                    new_display += guess
                    guessed_letters.append(guess)
                else:
                    new_display += display[i] if i < len(display) else "_"
            display = new_display
            print(display)

        elif guess not in chosen_word:
            lives -= 1
            print(guess, "is not in the word")
            print(stages[lives]) #todo 2
            guessed_letters.append(guess)
            print(display)

        if lives == 0:
            print(f"***********************YOU LOSE**********************")
            print(chosen_word)
            break


        if "_" not in display:
            print("****************************YOU WIN****************************")
            print("Word to guess: " + display)
            break



    #print(stages[lives])
```
